```python
from docx import Document
import numpy as np
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
import random

def extract_text_from_docx(file_path):
    doc = Document(file_path)
    text = ""  # 初始化为空字符串
    for para in doc.paragraphs:
        text += para.text + " "  # 将每个段落的文本追加到字符串中，并添加空格分隔
    return text

# 指定Word文件路径
file_path = r'Minutes of the Federal Open Market Committee June17-18, 2025.docx'
text = extract_text_from_docx(file_path)

# 指定中文字体路径（例如使用微软雅黑字体）
font_path = r'C:\Windows\Fonts\msyh.ttc'  # 微软雅黑字体路径

# 停用词设置
stopwords = set(STOPWORDS)
stopwords.update([
    "Federal", "Committee", "Meeting", "session",
    # 新增机构/会议相关术语
    "FOMC", "Reserve", "Desk", "participants", "staff", 
    "Board", "Governors", "System", "account",
    # 通用政策术语
    "policy", "monetary", "economic", "outlook", 
    "risks", "uncertainty", "objectives",
    # 高频动词/连接词
    "noted", "observed", "remarked", "commented", 
    "judged", "discussed", "agreed", "continued",
    # 时间/计量单位
    "percent", "basis", "points", "month", "quarter", 
    "year", "period", "intermeeting",
    # 通用经济术语
    "growth", "conditions", "developments", "indicators",
    # 文档特有关键冗余词
    "Projections", "summary", "review", "assessment"
])

# 生成词云图
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    font_path=font_path,  
    # color_func=custom_color_func,
    stopwords=stopwords,
    colormap='Set1', # tab20
).generate(text)

# 展示词云图
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
